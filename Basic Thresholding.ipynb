```python
import cv2
import numpy as np
```

```python
def show(label, image):
    cv2.imshow(label, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
img = cv2.imread("images/black_to_white.jpeg", cv2.IMREAD_GRAYSCALE)
```

```python
_, threshold_binary = cv2.threshold(img, 50, 255, cv2.THRESH_BINARY)
show("binary", threshold_binary)
```

```python
_, threshold_binary_inv = cv2.threshold(img, 50, 255, cv2.THRESH_BINARY_INV)
show("binary inv", threshold_binary_inv)
```

```python
_, threshold_trunc = cv2.threshold(img, 150, 255, cv2.THRESH_TRUNC)
show("trunc thresh", threshold_trunc)
# all above 150 are kept 150
```

```python
_, threshold_to_0 = cv2.threshold(img, 150, 255, cv2.THRESH_TOZERO)
show("thresh to 0", threshold_to_0)
# all below 150 are changed to 0
```
