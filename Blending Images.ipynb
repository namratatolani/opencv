```python
import cv2
import numpy as np
```

```python
road = cv2.imread("images/road.jpg")
car = cv2.imread("images/car.jpg")
car_gray = cv2.cvtColor(car, cv2.COLOR_BGR2GRAY)
ret, mask = cv2.threshold(car_gray, 252, 255, cv2.THRESH_BINARY_INV)
sum = cv2.add(car, car, mask = mask)
cv2.imshow("added masked", sum)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

now we will try to get the road background. the part not covered by the car.

```python
ret, mask = cv2.threshold(car_gray, 252, 255, cv2.THRESH_BINARY)
road_in_background = cv2.bitwise_and(road, road, mask=mask)
cv2.imshow("Road Background", road_in_background)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
mask_inv = cv2.bitwise_not(mask)
car_no_background = cv2.bitwise_and(car, car, mask=mask_inv)
cv2.imshow("Inverse Mask", mask_inv)
cv2.imshow("Car Background", car_no_background)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
result = cv2.add(road_in_background, car_no_background)
cv2.imshow("result", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Its not that great. But we can change thresholds and check.

```python
ret, mask = cv2.threshold(car_gray, 240, 255, cv2.THRESH_BINARY)
mask_inv = cv2.bitwise_not(mask)
car_no_background = cv2.bitwise_and(car, car, mask=mask_inv)
road_in_background = cv2.bitwise_and(road, road, mask=mask)
result = cv2.add(road_in_background, car_no_background)
cv2.imshow("result", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
