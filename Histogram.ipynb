Histogram gives intensity for each color in an image.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from commons import show
```

```python
img = np.zeros((100,100), np.uint8)
show("img", img)
```

```python
plt.hist(img.ravel(), 256, [0,256])
plt.show()
```

```python
cv2.rectangle(img, (0,50), (100,100), (255), -1)
show("img", img)
```

```python
plt.hist(img.ravel(), 256, [0,256])
plt.show()
```

```python
cv2.circle(img, (50, 50), 25, 127, thickness=-1)
show("img", img)
```

```python
plt.hist(img.ravel(), 256, [0,256])
plt.show()
```

```python
img = cv2.imread("images/sea.jpg", cv2.IMREAD_GRAYSCALE)
show("img", img)
```

```python
plt.hist(img.ravel(), 256, [0,256])
plt.show()
```

```python
img = cv2.imread("images/sea_beach.jpg")
b, g, r = cv2.split(img)
```

```python
cv2.imshow("img", img)
cv2.imshow("b", b)
cv2.imshow("g", g)
show("r", r)
```

```python
plt.hist(b.ravel(), 256, [0,256])
plt.hist(g.ravel(), 256, [0,256])
plt.hist(r.ravel(), 256, [0,256])
plt.show()
```
